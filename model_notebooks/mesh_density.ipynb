```python
densities_3 = [800000, 1223074, 2522258 , 4034587, 4173103, 4753697, 13561202, 14150302]
t2_ref_wind_speeds_3 =[7.12731, 7.00089, 6.95899, 6.971946, 6.9089236, 6.90566, 6.8946, 6.8951]
t3_ref_wind_speeds_3 = [7.19359, 6.942148, 7.359879, 7.380666, 7.34682, 7.39294, 7.353839, 7.363451]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = plt.axes()
plt.figure(dpi=1200)
plt.ticklabel_format(style='sci', axis='x', useMathText=True, useLocale=True)

plt.plot(densities_3, t2_ref_wind_speeds_3, label='T2')
plt.plot(densities_3, t3_ref_wind_speeds_3, label='T3')
plt.grid()
plt.legend()
plt.xlabel('Mesh Density')
plt.ylabel('Wind Speed (m/s)')
```

```python
dist = 10
a = plt.axes()
plt.figure(dpi=1200)
for dens, i in zip (densities_3, range(1,7)):
            #print(dens, dist, i)
    data = pd.read_csv('../mesh_density_files/3_turbine_%s_%sD.csv' %(i, dist))
    data= data.dropna()
    data['arc_length']=data['arc_length']-150
    plt.step(data['arc_length'], data['V_Magnitude'], label='%s cells, %sD distance'%(dens, dist))
plt.legend(loc=4, fontsize= '7')
plt.grid()
plt.xlabel('Distance to wind turbine hub (m)')
plt.ylabel('Wind Speed (m/s)')
```
